# Q1. Pearson Correlation Coefficient:
Suppose you have the following data on the amount of time students spend studying for an exam (in hours) and their final exam scores (out of 100):

python
Copy code
import numpy as np

    # Sample data
study_time = [10, 20, 15, 25, 30]
exam_scores = [60, 75, 70, 85, 90]

    # Calculate Pearson correlation coefficient
pearson_corr = np.corrcoef(study_time, exam_scores)[0, 1]

print(f"Pearson Correlation Coefficient: {pearson_corr:.2f}")
Interpretation:

If the Pearson correlation coefficient is close to 1, it indicates a strong positive linear relationship between study time and exam scores.
If it is close to -1, it indicates a strong negative linear relationship.
If it is close to 0, it suggests a weak or no linear relationship.
 # Q2. Spearman's Rank Correlation:
Suppose you have the following data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10:

python
Copy code
import scipy.stats as stats

    # Sample data
sleep_duration = [7, 8, 6, 5, 8]
job_satisfaction = [8, 7, 6, 4, 9]
 
      # Calculate Spearman's rank correlation
spearman_corr, _ = stats.spearmanr(sleep_duration, job_satisfaction)

print(f"Spearman's Rank Correlation: {spearman_corr:.2f}")
Interpretation:

A positive Spearman's rank correlation suggests a monotonic increase (or decrease) in job satisfaction with an increase in sleep duration.
A negative value indicates a monotonic decrease.
A value close to 0 implies a weak or no monotonic relationship.
# Q3. Comparison of Pearson and Spearman:
Suppose you have data on the number of hours of exercise per week and body mass index (BMI) for 50 participants:

python
Copy code
     # Assume you have the data in the variables exercise_hours and bmi

    # Calculate Pearson correlation coefficient
pearson_corr_exercise_bmi = np.corrcoef(exercise_hours, bmi)[0, 1]

    # Calculate Spearman's rank correlation
spearman_corr_exercise_bmi, _ = stats.spearmanr(exercise_hours, bmi)

print(f"Pearson Correlation: {pearson_corr_exercise_bmi:.2f}")
print(f"Spearman's Rank Correlation: {spearman_corr_exercise_bmi:.2f}")
Comparison:

Pearson measures linear relationships, while Spearman captures monotonic relationships.
If the relationship is strictly linear, Pearson and Spearman will provide similar results.
If the relationship is non-linear, Spearman might capture it better.
